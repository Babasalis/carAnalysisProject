```python
#Import the pandas library to reas csv file and numpy for math calculations
import pandas as pd
import numpy as np
#Read csv file and save it to variable called car and print
car = pd.read_csv('Sport car price.csv')
print(car.head())
```

```python
#Check the dtypes of the columns and if there any missing values
print(car.info())
"""
Looking at the Dtype in info(), a couple columns are supposed to be numerical-
Engine Size, Horsepower, Torque, MPH Time, Price
Also we have some null values in Engine Size and Torque columns as shown in the non-null count
"""
```

```python
#Initially, i tried to change the dtypes of Engine Size, Horsepower, Torque, 0-60 MPH Time, Price, but I got an error
#Look into the unique values of the columns you want to change to figure out why the dtype couldnt change
print('0-60 MPH Time (seconds) unique data', car['0-60 MPH Time (seconds)'].unique())
print('Engine Size (L) unique data', car['Engine Size (L)'].unique())
print('Horsepower unique data', car['Horsepower'].unique())
print('Torque (lb-ft) unique data', car['Torque (lb-ft)'].unique())
print('Price (in USD) unique data', car['Price (in USD)'].unique())

"""
0-60 MPH Time (seconds) has < attached to one of its values, Engine Size (L) has a couple things wrong, so it will require additional cleaning before it can be changed, Horsepower has , and + attached to some values, Torque (lb-ft) will also need additional cleaning before it can be changed, Price (in USD) has , in almost all the values
"""
```

#### Clean values

```python
#Clean the values of 0-60 MPH Time (seconds), Price (in USD), Horsepower
car["0-60 MPH Time (seconds)"] = car["0-60 MPH Time (seconds)"].str.replace("<", "")
car["Price (in USD)"] = car["Price (in USD)"].str.replace(",", "")
car["Horsepower"] = car["Horsepower"].str.replace(r"[,+]", "", regex=True)
#For Engine Size and Torque, the values in those columns are inconsistent, so I have to look deeper into those columns to properly clean them
```

```python
#Change the dtypes of these columns. 'Engine size' and 'Torque' will be changed later after further cleaning
car["0-60 MPH Time (seconds)"] = car["0-60 MPH Time (seconds)"].astype(float)
car["Price (in USD)"] = car["Price (in USD)"].astype(int)
car["Horsepower"] = car["Horsepower"].astype(int)

#Check to make sure the dtypes are changed
print(car.dtypes)
```

#### Clean Torque column and change dtype

```python
#Look through the unique values of Torque to see which values need to be replaced
print(car["Torque (lb-ft)"].unique())
#There are a couple of values that need to be replaced- '-', nan, '0'. Also, some values have ',' '+' attached to them that need to be removed.
```

```python
#Look into the rows that contain '-', nan, '0' as a value for Torque
print(car[car["Torque (lb-ft)"] == '-'])
print(car[car["Torque (lb-ft)"].isna()])
print(car[car["Torque (lb-ft)"] == '0'])
"""
Looking at this filtered dataframe one row has '-' as a torque value, three rows have nan, and one row has 0.
Also, Tesla Roadster comes up a couple times.
"""
```

```python
#Look into the rows that have Tesla Roadster
print(car[(car["Car Make"] == "Tesla") & (car["Car Model"] == "Roadster")])
#The Torque values for this Car Model is a mess. 10,000+ and 7,376 are also very high numbers for torque values.
```

```python
#Look into Tesla Model S Plaid to see what value we can use to replace the NaN value
print(car[(car["Car Make"] == "Tesla") & (car["Car Model"] == "Model S Plaid")])
#1050 is the value for all the rows, so 1050 will replace NaN
```

```python
#Look into Maserati GranTurismo to see what value we can use to replace the NaN value
print(car[(car["Car Make"] == "Maserati") &
          (car["Car Model"] == "GranTurismo")])
#384 is the value for all the rows, so 384 will replace NaN. Index 878 row values seem to be different from the rest.
```

```python
#Remove ,+ from Torque values
car["Torque (lb-ft)"] = car["Torque (lb-ft)"].str.replace(r"[,+]", "", regex=True)

#Replace '-', '0', nan, '10000', and '7376' with '737' because that is the consistent torque value for Tesla Roadster
car.loc[(car["Torque (lb-ft)"].isin(['-', '0', '10000', '7376'])
         | car["Torque (lb-ft)"].isna())
        & (car["Car Model"] == "Roadster"), "Torque (lb-ft)"] = '737'

#Check if values were replaced
print(car[(car["Car Make"] == "Tesla") & (car["Car Model"] == "Roadster")])
```

```python
#Change the NaN torque value of Tesla Model S Plaid and Maserati GranTurismo to '1050' and '384'
car.loc[(car["Torque (lb-ft)"].isna()) & (car["Car Model"] == "Model S Plaid"), "Torque (lb-ft)"] = '1050'
car.loc[(car["Torque (lb-ft)"].isna()) & (car["Car Model"] == "GranTurismo"), "Torque (lb-ft)"] = '384'
#Check if the values changed. This DataFrame should be empty now.
print(car[car["Torque (lb-ft)"].isna()])
#Since the values in Torque are now cleaned, we can go ahead and change the dtype
```

```python
#Convert Torque dtype to int
car["Torque (lb-ft)"] = car["Torque (lb-ft)"].astype(int)
#Check if Torque dtype was changed
print(car["Torque (lb-ft)"].dtypes)
```

#### Clean Engine Size column and change dtypes

```python
#Look through the unique values of Engine Size to see which values need to be replaced
print(car['Engine Size (L)'].unique())
"""
It is a whole mess. Numeric values, String values, '0', nan, '-' all in this column.
Engine Size should be a numeric column for calculations.
Electric Cars don't have engines, so what should I replace it with? You can't replace it with 0 because it will affect the math calculations.
"""
```

```python
#Look into the rows that contain '0', '-', and NaN as a value for Engine Size
print(car[car["Engine Size (L)"].isna()])
print(car[car["Engine Size (L)"].isin(['0', '-'])])
#All the rows are Electric cars.
```

```python
#Change '-', '0', and NaN values to "Electric" in Engine Size column
car.loc[(car["Engine Size (L)"].isin(['-', '0'])
         | car["Engine Size (L)"].isna())
        , "Engine Size (L)"] = 'Electric'
#Check to make sure these dataframes are now empty
print(car[car["Engine Size (L)"].isna()])
print(car[car["Engine Size (L)"].isin(["-", "0"])])
```

```python
#Look into the rows that contain Electric as a value
print(car[car["Engine Size (L)"].isin(["Electric (100 kWh)", "Electric", "Electric Motor", "2.0 (Electric)"])])
print(car[car["Engine Size (L)"].isin(["Electric (93 kWh)", "1.5 + Electric", "Electric (tri-motor)"])])
"""
It states in index 878 that the Maserati GranTurismo has an Electric engine type, but I don't think that is true, so i have to investigate further
With 'Electric (100 kWh)', 'Electric Motor', 'Electric (93 kWh)', and 'Electric (tri-motor)', i will go ahead and change the values to 'Electric' for consistency.
The 'BMW i8' is actually a Hybrid car with 1.5 engine size. Look into similar rows to find consistency before making any changes
'2.0 (Electric)' only has one row and it is associated with an electric car. Look into porsche taycan to make sure the value are consistent
"""
```

```python
#Look into rows with Maserati GranTurismo
print(car[(car['Car Make'] == 'Maserati') & (car['Car Model'] == 'GranTurismo')])
#Only one row has 'Electric' value in Engine Size column- index 878, so it has to be changed. Index 878 seems to have wrong values in other columns, but we will focus on Engine Size
```

```python
#Replace 'Electric' value with '4.7' for GranTurismo Car Model
car.loc[(car["Engine Size (L)"] == "Electric") & (car["Car Model"] == "GranTurismo"), "Engine Size (L)"] = '4.7'
#Check to make sure the value was changed
print(car[(car['Car Make'] == 'Maserati') & (car['Car Model'] == 'GranTurismo')])
```

```python
#Look into the Porsche Taycan rows
print(car[(car["Car Make"] == "Porsche") & (car["Car Model"] == "Taycan")])
#Porsche Taycan is an electric car. '2' and '2.0 (Electric)' are incorrect values and need to be changed
```

```python
#Replace '2' and '2.0 (Electric)' to 'Electric'
car.loc[(car['Engine Size (L)'].isin(['2', '2.0 (Electric)'])) &
        (car["Car Model"] == "Taycan"), 'Engine Size (L)'] = 'Electric'
#Check to make sure the values are changed
print(car[(car["Car Make"] == "Porsche") & (car["Car Model"] == "Taycan")])
```

```python
#Change 'Electric (100 kWh)', 'Electric Motor', 'Electric (93 kWh)', and 'Electric (tri-motor)' to 'Electric'
car['Engine Size (L)'] = car['Engine Size (L)'].replace(
    ['Electric (100 kWh)', 'Electric Motor', 'Electric (93 kWh)', 'Electric (tri-motor)'], 'Electric')
#Check to make sure none of these values no longer exist in 'Engine Size (L)
print(car[car["Engine Size (L)"].isin(['Electric (100 kWh)', 'Electric Motor', 'Electric (93 kWh)', 'Electric (tri-motor)'])])
```

```python
#Look into the rows with BMW i8
print(car[(car['Car Make'] == 'BMW') & (car['Car Model'] == 'i8')])
#BMW i8 is a hybrid car with 1.5 Engine Size
```

```python
#Change the values of the Engine Size column related to BMW i8 to 1.5
car.loc[(car['Engine Size (L)'].isin(['1.5 + Electric', 'Hybrid'])) &
        (car['Car Model'] == 'i8'), 'Engine Size (L)'] = '1.5'
#Check to make sure the values were changed
print(car[(car['Car Make'] == 'BMW') & (car['Car Model'] == 'i8')])
```

```python
#Look into the rows that contain Hybrid as an Engine Size value
print(car[car["Engine Size (L)"].isin(["Hybrid", "Hybrid (4.0)", "4.0 (Hybrid)"])])
```

```python
#Look into Porsche Panamera Turbo S E-Hybrid and Ferrari SF90 Stradale
print(car[(car['Car Make'] == 'Porsche') & (car['Car Model'] == 'Panamera Turbo S E-Hybrid')].head())
print(car[(car['Car Make'] == 'Ferrari') & (car['Car Model'] == 'SF90 Stradale')].head())
"""
With the porsche Panamera Turbo S E-Hybrid, the engine size is 4.0
With Ferrari SF90 Stradale, the engine size is also 4.0
"""
```

```python
#Replace "Hybrid", "Hybrid (4.0)", "4.0 (Hybrid)" with "4.0"
car['Engine Size (L)'] = car['Engine Size (L)'].replace(['Hybrid', 'Hybrid (4.0)', '4.0 (Hybrid)'], '4.0')
#Make sure none of this values come up in Engine Size (L)
print(car[car["Engine Size (L)"].isin(["Hybrid", "Hybrid (4.0)", "4.0 (Hybrid)"])])
```

#### Create two new columns-'Engine Size(Liters)' and 'Car Type'- to save engine size and the type of car

```python
#This code will turn all the non-numerical values to NaN, and store the values in Engine Size(Liters). This code normally won't run, but because of errors = 'coerce' as an argument, it turns non-numeric data into NaN values
car["Engine Size(Liters)"] = pd.to_numeric(car["Engine Size (L)"], errors='coerce')

#This code will store the values of Engine Size (L) where Engine Size(Liters) is null, so all the string values of Engine Size (L) will be saved here
car["Car Type"] = car["Engine Size (L)"].where(car["Engine Size(Liters)"].isna())
#Change all the NaN values in car type to Gas
car["Car Type"] = car["Car Type"].replace(np.nan, 'Gas')

#Check to make sure these two columns have been created and also the dtypes of these columns
print(car[['Engine Size (L)', 'Engine Size(Liters)', 'Car Type']].tail())
print(car[['Engine Size(Liters)', 'Car Type']].dtypes)
```

#### Drop Engine Size (L) from the DataFrame

```python
"""
Engine Size (L) column is no longer needed in this dataset, so I will remove it.
Save the new Dataframe to a variable called edited_car to not temper with the original dataset, car.
"""
#Drop Engine Size (L)
edited_car = car.drop('Engine Size (L)', axis=1)
#Check edited_car to make sure 'Engine Size (L)' no longer exist
print(edited_car.head())
```

#### Save edited_car to a csv file

```python
edited_car.to_csv('edited.csv')
```
